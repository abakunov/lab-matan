```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Function declaration
function = lambda n: (5*n-7)/(2*n+5)*math.cos(((2+(-1)**n)/6)*math.pi) 
even_function = lambda n: (5*n-7)/(2*n+5)*math.cos(((2+1)/6)*math.pi)
odd_function = lambda n: (5*n-7)/(2*n+5)*math.cos(((2-1)/6)*math.pi)

#Setting the values obtained during the analytical method
SUP = 5*math.sqrt(3)/4
INF = -math.sqrt(3)/7
LIM_SUP = SUP
LIM_INF = 0

#Setting the number of points
AMOUNT_OF_POINTS = 50
#Set EPSILON value here!
EPSILON = 0.5

#Calculating the m_sup value
def check_criterion(function, epsilon, SUP):
    m_sup = 1
    while True:
        if function(m_sup) > SUP - epsilon:
            break
        m_sup += 1
    return m_sup

m = check_criterion(function, EPSILON, SUP)
print("The value of m is: ", m)

AMOUNT_OF_POINTS = max(AMOUNT_OF_POINTS, m)

#Creating the plot
plt.plot(np.arange(1, AMOUNT_OF_POINTS), [even_function(n) for n in np.arange(1, AMOUNT_OF_POINTS)], 'o', color='#32c7f0', label='Even n', markersize=3)
plt.plot(np.arange(1, AMOUNT_OF_POINTS), [odd_function(n) for n in np.arange(1, AMOUNT_OF_POINTS)], 'o', color='#0e993c', label='Odd n', markersize=3)
plt.plot(np.arange(1, AMOUNT_OF_POINTS), [SUP]*(AMOUNT_OF_POINTS-1), color='#ab3f95', label='Sup')
plt.plot(np.arange(1, AMOUNT_OF_POINTS), [INF]*(AMOUNT_OF_POINTS-1), color='#151d75', label='Inf')
plt.plot(np.arange(1, AMOUNT_OF_POINTS), [LIM_INF]*(AMOUNT_OF_POINTS-1), color='#990e45', label='Lower limit')
plt.plot(m, function(m), 'o', color='#ff0000', label='m')
plt.title('Tasks 1, 2, 4')
plt.legend()
```

```python
#Checking the epsilon neighborhood
def epsilon_neighborhood_ckecker(f, n, epsilon):
    return abs(f(n)-LIM_SUP) < epsilon

#Finding the first n that satisfies the epsilon neighborhood
n_0 = 1
while True:
    if epsilon_neighborhood_ckecker(odd_function, n_0, EPSILON):
        break
    n_0 += 1

print("The first n that satisfies the epsilon neighborhood condition is: ", n_0)

#Creating the plot
plt.plot(np.arange(n_0, n_0+AMOUNT_OF_POINTS), [LIM_SUP]*(AMOUNT_OF_POINTS), color='green', label='Upper limit')
plt.plot(np.arange(n_0, n_0+AMOUNT_OF_POINTS), [odd_function(n) for n in np.arange(n_0, n_0+AMOUNT_OF_POINTS)], 'o', color='#896790', label='Odd n', markersize=3)
plt.title('Task 3')
plt.legend()
```
